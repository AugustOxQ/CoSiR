```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import json
import os
from PIL import Image
```

```python
path = "/data/SSD2/pre_extract/impressions/sample_ids.pt"
```

```python
# load the sample_ids.pt file
sample_ids = torch.load(path)

# print the sample_ids
print(sample_ids[:10])
```

```python
data_path = "/project/Impressions/metadata/impressions_train.json"
image_path = "/project/Impressions/media"
data = json.load(open(data_path))

real_idx = sample_ids[:5]
print(data[real_idx[1]])
```

```python
# Plot in subplots
fig, axs = plt.subplots(1, 5, figsize=(20, 5))
for i, idx in enumerate(real_idx):
    image = os.path.join(image_path, data[idx]["image"])
    image = Image.open(image)
    axs[i].imshow(image)
    axs[i].set_title(data[idx]["caption_type"])
plt.show()
```

```python
captions_list = [data[idx]["caption"] for idx in real_idx]
print(*captions_list, sep="\n")
```
